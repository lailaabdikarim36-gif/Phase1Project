# Aviation Risk Analysis

## Project Overview

The purpose of this project is to analyze historical aviation accident data to assess the level of risk associated with different aircraft types. The analysis is conducted to support a business decision for a company that is planning to expand into the aviation industry by purchasing and operating aircraft for commercial and private use.

Using data-driven insights from past aviation accidents, this project aims to identify aircraft that present the lowest operational risk. The findings will be translated into clear, actionable recommendations for non-technical business stakeholders.


## Business Understanding

The company is seeking to diversify its portfolio by entering the aviation industry. However, leadership lacks experience in aviation operations and needs guidance on which aircraft are safest to purchase.

The key business objective of this analysis is to determine which aircraft makes and models have historically demonstrated the lowest risk based on accident frequency and severity.

The primary stakeholder for this project is the head of the new aviation division, who will use these insights to make informed aircraft purchasing decisions that minimize financial, operational, and human risk.


## Data Understanding

The dataset used in this analysis comes from the National Transportation Safety Board (NTSB) and contains records of civil aviation accidents and selected incidents from 1962 to 2023 in the United States and international waters.

The dataset includes information such as:
- Event date and location
- Aircraft make and model
- Aircraft category and engine type
- Injury severity and number of fatalities
- Weather conditions and phase of flight
- Aircraft damage level

This dataset provides a comprehensive historical view of aviation accidents, making it suitable for identifying patterns related to aircraft safety and operational risk.

In this section, the aviation accident dataset is loaded into a pandas DataFrame. This allows for efficient data exploration, cleaning, aggregation, and analysis throughout the project.


```python
import pandas as pd
import numpy as np

# Load the aviation accident dataset
df = pd.read_csv("data/Aviation_Data.csv" ,low_memory = False)

# Display the first few rows of the dataset
df.head()
```

## Data Preparation

In this section, relevant columns are selected and prepared for analysis. Columns with excessive missing values or those not relevant to assessing aircraft risk are excluded to ensure reliable and interpretable results.


```python
columns_of_interest = [
    "Event.Date",
    "Make",
    "Model",
    "Injury.Severity",
    "Aircraft.damage",
    "Weather.Condition",
    "Number.of.Engines",
    "Total.Fatal.Injuries",
    "Total.Serious.Injuries",
    "Total.Minor.Injuries",
    "Total.Uninjured"
]
df_clean = df[columns_of_interest]
```

```python
df_clean.head()
```

```python
# Avoid chained assignment issues
df_clean = df[columns_of_interest].copy()
```

```python
df_clean.isna().sum()
```

## Data Cleaning and Standardization

In this section, the dataset is cleaned and standardized to improve consistency and reliability for analysis.  
Text fields are formatted to remove extra spaces and unify capitalization, and missing or invalid values are handled to reduce noise in the dataset.

### What we do
- **Standardize text columns** (e.g., Make, Model, Weather condition) by stripping whitespace and applying consistent capitalization.
- **Handle missing values** in injury-related columns by replacing nulls with `0`, since a missing entry likely indicates no injuries were recorded.
- **Convert common placeholders** such as `"UNK"`, `"Unknown"`, `"N/A"`, or empty strings into proper missing values (`NaN`) so they can be treated correctly.
- **Drop rows with missing critical information** (such as aircraft make/model or weather condition) because these fields are essential for meaningful analysis.

After these steps, the dataset becomes cleaner, more uniform, and ready for feature engineering and visualization.


```python
# Standardization
text_cols = ['Make', 'Model', 'Weather.Condition']

for col in text_cols:
    df_clean.loc[:, col] = df_clean[col].str.strip().str.title()

    
```

```python
# Data Cleaning
injury_cols = [
    'Total.Fatal.Injuries',
    'Total.Serious.Injuries',
    'Total.Minor.Injuries',
    'Total.Uninjured'
]

# Fill missing injury values with 0
df_clean[injury_cols] = df_clean[injury_cols].fillna(0)

# Replace unknown placeholders with NaN (text columns only)
unknown_values = ['UNK', 'Unknown', 'N/A', '']
df_clean[text_cols] = df_clean[text_cols].replace(unknown_values, pd.NA)

# Drop rows missing critical information
df_clean = df_clean.dropna(
    subset=['Make', 'Model', 'Weather.Condition', 'Number.of.Engines']
)
```

```python
# Fill remaining text NaNs 
text_cols = ['Injury.Severity', 'Aircraft.damage']
df_clean[text_cols] = df_clean[text_cols].fillna('Unknown')
```

```python
df_clean.isna().sum()
```

## Feature Engineering

In this section, new features are created from existing variables to enhance the analytical value of the dataset.  
A severity score is computed to quantify the overall impact of each aviation incident based on injury outcomes.


### Creating a Severity Factor

A severity score is created to measure how serious each aviation accident was.

- Fatal injuries are weighted 5 times.
- Serious injuries are weighted 3 times.
- Minor injuries are weighted 1 time.

This score helps compare accidents based on severity.


```python
df_clean['Severity.Score'] = (
    5 * df_clean['Total.Fatal.Injuries'] +
    3 * df_clean['Total.Serious.Injuries'] +
    1 * df_clean['Total.Minor.Injuries']
)
```

```python
df_clean[['Severity.Score']].head()
```

```python
# Convert Event.Date to datetime format
df_clean["Event.Date"] = pd.to_datetime(df_clean["Event.Date"], errors="coerce")

# Extract year from the event date
df_clean["Year"] = df_clean["Event.Date"].dt.year

# Preview changes
df_clean.head()
```

## Data Analysis
The analysis aggregates accident data by aircraft manufacturer to evaluate relative risk. Metrics include total number of accidents, number of fatal accidents, and fatal accident rate. Manufacturers with fewer than 50 recorded accidents were excluded to ensure statistical reliability.


```python
# Create a binary indicator for fatal accidents
df_clean["Is_Fatal"] = df_clean["Total.Fatal.Injuries"].fillna(0) > 0

# Convert boolean to integer (1 = Fatal, 0 = Non-Fatal)
df_clean["Is_Fatal"] = df_clean["Is_Fatal"].astype(int)

# Preview the result
df_clean[["Injury.Severity", "Total.Fatal.Injuries","Is_Fatal"]].head()

```

```python
# Group by aircraft make and calculate risk metrics
make_risk = (
    df_clean
    .groupby("Make" )
    .agg(
        total_accidents=("Is_Fatal", "count"),
        fatal_accidents=("Is_Fatal", "sum"),
        fatal_rate=("Is_Fatal", "mean")
    )
    .reset_index()
)

make_risk.head()

```

```python
# Keep only manufacturers with enough data
make_risk_filtered = make_risk[make_risk["total_accidents"] >= 50]
 
# Sort by fatal rate (lowest risk first)
make_risk_filtered = make_risk_filtered.sort_values("fatal_rate")

make_risk_filtered.head(10)
```

```python
top10_safe_makes = make_risk_filtered.head(10)
top10_safe_makes
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
plt.barh(
    top10_safe_makes["Make"],
    top10_safe_makes["fatal_rate"]
)
plt.xlabel("Fatal Accident Rate")
plt.ylabel("Aircraft Manufacturer")
plt.title("Lowest Fatal Accident Rates by Aircraft Manufacturer")
plt.gca().invert_yaxis()  # Lowest risk at the top
plt.show()
```

## Conclusion
Based on historical accident data, several aircraft manufacturers demonstrate lower fatal accident rates compared to others.
